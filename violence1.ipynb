```python
import os
import cv2
from tqdm import tqdm

# Provided paths
non_violence_path = "/kaggle/input/images/NonViolence/NonViolence"
violence_path = "/kaggle/input/images/Violence/Violence"

# List to store data and labels
data = []
labels = []

# Iterate through NonViolence images
for imagePath in tqdm(os.listdir(non_violence_path)[:5000]):
    image_path = os.path.join(non_violence_path, imagePath)
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (224, 224))
    data.append(image)
    labels.append(0)  # Assigning 0 for NonViolence

# Iterate through Violence images
for imagePath in tqdm(os.listdir(violence_path)[:7000]):
    image_path = os.path.join(violence_path, imagePath)
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (224, 224))
    data.append(image)
    labels.append(1)  # Assigning 1 for Violence
```

```python
data = np.array(data)
labels = np.array(labels)
```

```python
lb = LabelBinarizer()
labels = lb.fit_transform(labels)
labels = to_categorical(labels)
```

```python
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras.layers import AveragePooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.preprocessing.image import img_to_array
from collections import deque
from tensorflow.keras.layers import Dropout
```

```python
base_model = InceptionV3(weights="imagenet", include_top=False, input_tensor=Input(shape=(224, 224, 3)))
head_model = base_model.output
head_model = AveragePooling2D(pool_size=(5, 5))(head_model)
head_model = Flatten(name="flatten")(head_model)
head_model = Dense(512, activation="relu")(head_model)
head_model = Dropout(0.5)(head_model)
head_model = Dense(2, activation="softmax")(head_model)
```

```python
model = Model(inputs=base_model.input, outputs=head_model)
```

```python
model.trainable = True
```

```python
opt = SGD(learning_rate=1e-4, momentum=0.9, decay=1e-4 / args["epochs"])
model.compile(loss="categorical_crossentropy", optimizer=opt, metrics=["accuracy"])
print(model.summary())
```

```python
lstm_model = Sequential([
    LSTM(units=128, return_sequences=True, input_shape=(None, 2048)),  # Increased units and return sequences
    Dropout(0.5),  # Dropout for regularization
    LSTM(units=64),  # Additional LSTM layer for deeper architecture
    Dense(units=32, activation='relu'),  # Additional fully connected layer for feature extraction
    Dropout(0.5),  # Dropout for regularization
    Dense(units=1, activation='sigmoid')  # Output a binary classification (violence or non-violence)
])
```

```python
opt = SGD(lr=1e-4, momentum=0.9, decay=1e-4 / args["epochs"])
lstm_model.compile(loss="binary_crossentropy", optimizer=opt, metrics=["accuracy"])
print(lstm_model.summary())
```
